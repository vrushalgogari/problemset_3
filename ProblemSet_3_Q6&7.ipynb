```python
import pandas as pds
df = pds.read_csv(r"C:\Users\visha\Desktop\us-marriages-divorces-1867-2014.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

fig= plt.figure(figsize=(10,10))

df.plot('Year', y=['Marriages', 'Divorces'])
plt.show()
```

```python
import matplotlib.pyplot as plt

df1 = df[df['Year'] >= 1900 ]
df2 = df1[df['Year'] <=2000 ]
df2.head()

fig= plt.figure(figsize=(10,10))
plt.bar(df2['Marriages_per_1000'],df2['Divorces_per_1000'], label=" number of marriages and divorces per capita ")
plt.legend()

plt.xlabel('Marriages')
plt.ylabel('Divorces')
plt.title('Vertical Bar chart')

plt.show()
```
