```python
users = pds.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
users.head(20)
```

```python
# Select an occupation by groupby and then find the mean of the age column among them.
users[["age", "occupation"]].groupby("occupation").mean()
```

```python
def conv_gender_to_numeric(x):
    if x == 'M':
        return 1
    if x == 'F':
        return 0
    
users['gender_num'] = users['gender'].apply(conv_gender_to_numeric)

ratio = users.groupby('occupation').gender_num.sum() / users.occupation.value_counts() * 100 
# users.occupation.value_counts() 
ratio.sort_values(ascending = False)
```

```python

users.groupby('occupation').age.agg(['min', 'max'])
```

```python
users.groupby(['occupation', 'gender']).age.mean()
```

```python

# First, divide by occupation and gender and calculate the total number of males and females. 
a  =  users . groupby ([ 'occupation' , 'gender' ]) . agg ({ 'gender' : 'count' }) 
# And get the sum of each column by occupation = sum of gender 
b  =  users . groupby ([ 'occupation' ]) . agg ( 'count' ) 
# Divide a by b. Level makes the criterion occupation. Coming out decimal * 100 
c  =  a . div ( b , level = 'occupation' ) *  100 
# View only the gender column 
# Look at all rows with loc and select gender for the column 
c . loc [:, 'gender' ]
```
