```python
import pandas as pds
import datetime
```

```python
data = pds.read_csv(r"C:\Users\visha\Desktop\wind.txt", sep = "\s+", parse_dates = [[0,1,2]]) 
data.head()
```

```python
def fix_century(x):
  year = x.year - 100 if x.year > 1989 else x.year
  return datetime.date(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_century)

data.head()
```

```python
data["Yr_Mo_Dy"] = pds.to_datetime(data["Yr_Mo_Dy"])

data = data.set_index('Yr_Mo_Dy')

data.head()
```

```python
day_stats = pds.DataFrame()

day_stats['minimum'] = data.min(axis = 1) # minimum
day_stats['maximum'] = data.max(axis = 1) # maximum 
day_stats['mean'] = data.mean(axis = 1) # mean value
day_stats['std_dev'] = data.std(axis = 1) # standard deviations

day_stats.head()
```
